```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
merge_df = pd.merge(city_df, ride_df, on="city")

# Display the data table for preview
```

```python
merge_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
urban_city = merge_df.loc[(merge_df["type"] == "Urban")]
suburban_city = merge_df.loc[(merge_df["type"] == "Suburban")]
rural_city = merge_df.loc[(merge_df["type"] == "Rural")]

#urban
urban = urban_city.groupby(['city'])
#Average Fare ($) Per City
avg_fare_urban = urban["fare"].mean()
#Total Number of Rides Per City
number_rides_urban = urban["ride_id"].count()
#Total Number of Drivers Per City
drivercount_urban = urban["driver_count"].sum()
drivers_urban = (drivercount_urban // number_rides_urban)

#suburban
suburban = suburban_city.groupby(['city'])
#Average Fare ($) Per City
avg_fare_suburban = suburban["fare"].mean()
#Total Number of Rides Per City
number_rides_suburban = suburban["ride_id"].count()
#Total Number of Drivers Per City
drivercount_suburban = suburban["driver_count"].sum()
drivers_suburban = (drivercount_suburban // number_rides_suburban)

#rural
rural = rural_city.groupby(['city'])
#Average Fare ($) Per City
avg_fare_rural = rural["fare"].mean()
#Total Number of Rides Per City
number_rides_rural = rural["ride_id"].count()
#Total Number of Drivers Per City
drivercount_rural = rural["driver_count"].sum()
drivers_rural = (drivercount_rural // number_rides_rural)

```

```python
# Create scatter plots for Urban, Suburban and Rural cities
plt.scatter(number_rides_urban, avg_fare_urban ,s=number_rides_urban*10,
            marker ='o', facecolors ="lightcoral", edgecolors='black', label="Urban")

plt.scatter(number_rides_suburban, avg_fare_suburban ,s=number_rides_suburban*10,
            marker ='o', facecolors ="lightskyblue", edgecolors='black', label="Suburban")

plt.scatter(number_rides_rural, avg_fare_rural ,s=number_rides_rural*10,
            marker ='o', facecolors ="gold", edgecolors='black', label="Rural")


# Chart title
plt.title("Pyber Ride Sharing Data (2016)")
# x label
plt.xlabel("Total Numbers of Rides (Per City))")
# y label
plt.ylabel("Average Fare($)")
# legend 
plt.legend(loc='upper right')

plt.text(45, 30,"Note: Circle size correlates with driver count per city.")
plt.grid()
plt.savefig("../Pyber Ride Sharing.png")
# Incorporate a text label regarding circle size
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fare_type = merge_df.groupby(["type"])['fare'].sum()

# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]


# Build Pie Chart
plt.title("Total Fares By City Type")
plt.pie(fare_type, labels=labels, colors=colors,
         autopct="%1.1f%%", shadow=True, startangle=0)
plt.axis("equal")

# Save Figure
plt.savefig("../fare pie.png")
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ride_type = merge_df.groupby(["type"])['ride_id'].count()

# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]

# Build Pie Chart
plt.title("Total Rides By City Type")
plt.pie(ride_type, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=0)
plt.axis("equal")

# Save Figure
plt.savefig("../ride pie.png")
# Show Figure
plt.show()

```

## Total Drivers by City Type

```python
# Calculate Driver Percents
total_drivers = merge_df.groupby(["type"])['driver_count'].sum()

pie_drivers = (total_drivers // ride_type)

# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]

# Build Pie Chart
plt.title("Total Drivers By City Type")
plt.pie(pie_drivers, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=0)
plt.axis("equal")

# Save Figure
plt.savefig("../driver pie.png")
# Show Figure
plt.show()
```
